```python
import matplotlib.pyplot as plt
```

```python
categories = ['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
```

```python
revenue = [25,15,5,20,10,25]
plt.pie(revenue)
plt.show()
```

```python
#explode the shirts piece to make more noticeable
explode = [0,.1,0,0,0,0]
```

```python
#plt.pie(revenue,label=categories)
plt.pie(revenue,labels=categories,autopct="%1.1f%%",explode=explode)
plt.show()
```
